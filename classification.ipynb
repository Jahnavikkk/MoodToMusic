# Import packages

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import cv2
```

```python
TRAIN_DIR = "FER13/train"
TEST_DIR = "FER13/test"
```

<h1>EDA</h1>

```python
classes = os.listdir(TRAIN_DIR)
classes
```

```python
image_paths = [TRAIN_DIR+"/"+c+"/"+os.listdir(os.path.join(TRAIN_DIR,c))[0] for c in classes]
images = [cv2.imread(path,0) for path in image_paths]
```

```python
fig,ax = plt.subplots(1,len(images),figsize=(20,10))
for ii,c_ax in enumerate(ax):
    c_ax.imshow(images[ii],cmap="gray")
    c_ax.set_title(classes[ii])
```
